**Overall Key Insights**

<u>Data Completeness and Quality:</u>

The dataset has missing values in the project completion date and planned tax benefit columns. There was also a mismatch in the provided data dictionary that came with the dataset and the actual variables found in the dataset. For example, the program group column is in the dataset, but not the data dictionary. Bedroom count columns are in the data dictionary, but not in the dataset, etc.

An additional variable was created to focus on completed projects and their duration. Two rows show a negative value for project duration, revealing some administrative errors when the project start and completion dates were being entered.

After data cleaning, 2658 projects were shown to be completed out of the 3634 projects listed in the initial data. 

<u>Variables:</u> 

Distribution Skewness:

Distributions of numerical columns are generally skewed to the right.

On average, the largest projects were very low-income units among the different income unit categories, while moderate-income units had the lowest mean. Low_income_units are the second highest.

The mean for counted rental units was higher than counted homeownership units.

The 'Multifamily Incentives Program' is the most frequent category in the 'program_group' column, with ‘Confidential’ as second.

Project Duration and Size:

The average project took 281 days to complete, with some projects taking over six years! The median project took 135 days.

Projects under the Small Homes Program have the longest average duration, while those under the Multifamily Incentives Program have the shortest.

The smallest project had a total of one unit, while some of the largest projects were very low-income and homeownership.

<u>Overall Conclusion:</u>

The analysis shows some interesting connections between program types, project volume, and affordability percentages. In Christina's bar plot of Program Group, we see that the Multi-Family Incentives Program has the most projects, while the Small Homes Program has the fewest. This lines up with the affordability breakdown in Angeneris' distribution of affordable housing percentage per building and Josue's distribution of affordable percentage graph, where most buildings fall into the 30% and 100% affordability categories. It seems that larger programs like Multi-Family Incentives are focusing on these affordability levels to meet certain funding or policy goals. On the other hand, the Small Homes Program, with fewer projects and longer durations, might be tackling more complex or individualized builds. These projects could face additional challenges, like zoning issues or land use restrictions, which could explain the longer timelines even though the homes are smaller.

The concentration of projects at 100% and 30% affordability also suggests that there's a strong focus on providing affordable housing at these levels, possibly due to increased resources or policy attention. The contrast between Multi-Family Incentives and Small Homes highlights a trade-off between project volume and duration: larger programs benefit from economies of scale, leading to quicker build times, while smaller programs face unique challenges that extend their timelines. This all points to the complex relationship between program size, affordability goals, and project duration.
